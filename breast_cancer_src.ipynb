```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading data

```python
data = load_breast_cancer()
df_features = pd.DataFrame(data=data.data, columns=data.feature_names).iloc[:, :10]
df_label = pd.DataFrame(data=data.target, columns=['label'])
df = pd.concat([df_features, df_label], axis=1)
df.head()
```

```python
df_features.head()
```

```python
df_label.head()
```

```python
sns.pairplot(df, hue='label')
```

## Normalising

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df_features_normalized = pd.DataFrame(data=MinMaxScaler().fit_transform(df_features), columns=df_features.columns)
df_features_normalized
```

```python
df_features_normalized.describe()
```

## Principal Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(df_features_normalized)
per_var = np.round(pca.explained_variance_ratio_, decimals=1)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(per_var)+1), per_var.cumsum(), marker = "o", linestyle = "--")
plt.grid()
plt.ylabel("Percentage Cumulative of Explained Variance")
plt.xlabel("Number of Components")
plt.title("Explained Variance by Component")
plt.show()
```

```python
pca = PCA(n_components=3)
pca.fit(df_features_normalized)
df_pca = pca.transform(df_features_normalized)
```

```python
df_pca
```

## K-means++

```python
from sklearn.cluster import KMeans
```

```python
dist_to_centroid = []

for i in range(1, 30):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df_pca)
    dist_to_centroid.append(kmeans.inertia_)
```

```python
plt.figure(figsize = (10,6))
plt.plot(range(1,30), dist_to_centroid, marker = "o", linestyle = "--")
plt.grid()
plt.title("Cluster using PCA Scores")
plt.ylabel("Within Cluster Sum of Squares")
plt.xlabel("N Clusters")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
kmeans.fit(df_pca)
```

```python
df_kmeans = pd.concat([df_features, pd.DataFrame(df_pca)], axis=1)
df_kmeans.columns.values[-3:] = ['comp1', 'comp2', 'comp3']
df_kmeans['segment'] = kmeans.labels_
df_kmeans['segment'] = df_kmeans['segment'].map({0:'Cluster 1', 1:'Cluster 2', 2:'Cluster 3', 3:'Cluster 4', 4:'Cluster 5'})
df_kmeans.head()

df1 = df_kmeans.iloc[:, 10:14]
df1.head()
```

```python
sns.pairplot(df1, hue='segment')
```

```python
from sklearn.metrics import silhouette_score

s_score = silhouette_score(df1.iloc[:, :-1], df1.iloc[:, -1])
print(s_score)
```

## DBSCAN

```python
min_points = 6
```

```python
from sklearn.neighbors import NearestNeighbors

nn = NearestNeighbors(n_neighbors=min_points - 1)
neighbours = nn.fit(df_pca)
distances, indices = neighbours.kneighbors(df_pca)

shortest_distance = np.sort(distances, axis=0)[:, 1]
print(shortest_distance)
```

```python
plt.figure(figsize=(12,8))
plt.plot(shortest_distance)
plt.xlabel('Data points sorted by distance')
plt.ylabel('Epsilon value')
```

```python
epsilon = 0.125
```

```python
from sklearn.cluster import DBSCAN

dbs = DBSCAN(eps=epsilon, min_samples=min_points)
dbs.fit(df_pca)
```

```python
df_dbs = pd.DataFrame(df_pca).copy()
df_dbs['segment'] = dbs.labels_
df_dbs.head()
```

```python
df_dbs.describe()
```

```python
s_score = silhouette_score(df_dbs, dbs.labels_)
print(s_score)
```
